```python
#Question3: What is App routing in Flask? Why do we use app routes?



"""In Flask, app routing refers to the process of mapping a URL to a specific function in your Flask application. Flask uses 
decorators to define routes, which are functions that are called when a specific URL is 
requested.

For example, if you want to create a route for the homepage of your application, you can use the @app.route() decorator to 
define a function that will be called when a user requests the URL associated with the route:
    
from flask import Flask

app = Flask(__name__)

@app.route('/')
def index():
    return 'Hello, World!'


The main benefits of using app routing in Flask are:

Clear and organized code: App routing allows you to organize your code into functions that handle specific requests, making it 
    easier to maintain and modify your code as your application grows.

Separation of concerns: By separating the logic of your application into different routes, you can keep the different parts of 
    your application organized and focused on their specific responsibilities.

Better user experience: With app routing, you can create dynamic web applications with multiple pages and functionality, which 
    can provide a better user experience by allowing users to easily 
    navigate through your application and access the content they need."""
```

```python
#Question5:  What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the
# url_for() function.




n Flask, the url_for() function is used for URL building. The url_for() function generates a URL for the specified function and
arguments. The function takes the name of the function as its first argument and any additional arguments that the function 
requires.

Here's an example of how to use the url_for() function in Flask:


from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def index():
    return 'Hello, World!'

@app.route('/user/<username>')
def show_user_profile(username):
    return f'User {username}'

@app.route('/post/<int:post_id>')
def show_post(post_id):
    return f'Post {post_id}'

with app.test_request_context():
    print(url_for('index'))  # Output: '/'
    print(url_for('show_user_profile', username='John Doe'))  # Output: '/user/John%20Doe'
    print(url_for('show_post', post_id=1))  # Output: '/post/1'
```
